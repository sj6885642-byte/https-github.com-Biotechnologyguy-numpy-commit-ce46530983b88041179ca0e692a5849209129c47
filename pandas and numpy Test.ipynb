```python
print(df.head())
```

```python
avg_playtime = df.groupby("GameGenre")["PlayTimeHours"].mean()
print(avg_playtime)
```

```python
df["EngagementScore"] = df["EngagementLevel"].map({"Low":1,"Medium":2,"High":3})
avg_engagement = df.groupby("Location")["EngagementScore"].mean()
print(avg_engagement.idxmax())
```

```python
avg_purchases = df.groupby("Gender")["InGamePurchases"].mean()
print(avg_purchases)
```

```python
df["AgeGroup"] = pd.cut(df["Age"],bins=[0,20,30,40,100],labels=["Below 20","20-30","30-40","Above 40"])
avg_playtime = df.groupby("AgeGroup")["PlayTimeHours"].mean()
print(avg_playtime)
```

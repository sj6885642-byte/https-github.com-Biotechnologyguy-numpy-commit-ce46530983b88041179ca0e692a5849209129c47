```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\HP\Pictures\Education_Course_Institute_India_With_Fees.xlsx")
df.head()
```

```python
df["Year"] =df["Enrollment Date"].dt.year
df["Year"] =df["Year"].astype(str)
```

```python
df["Quarter"] =df["Enrollment Date"].dt.quarter
df["Quarter"].replace({1 :"Q1" ,2: "Q2" , 3 : "Q3" , 4 : "Q5"})
df
```

```python
plt.figure(figsize = (6,2))
sns.lineplot(x = "Year" , y = "Fees" , data = df , estimator = sum)
plt.title("Total Sales Over Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize = (6,2))
sns.lineplot(x = "Quarter" , y = "Fees" , data = df , estimator = len)
plt.title("Total Admissions By Quarter")
plt.xlabel("Quarter")
plt.ylabel("No.Of Students")
plt.show()
```

```python
df["Quarter"] = df["Enrollment Date"].dt.quarter
df["Quarter"].replace({1 :"Q1" ,2: "Q2" , 3: "Q3" , 4: "Q4"})
df
```

```python
plt.figure(figsize = (6,2))
sns.lineplot(x = "Quarter" , y = "Fees" , data = df , estimator = len , hue = "Placement Status")
plt.title("Total Admissions By Quarter")
plt.xlabel("Quarter")
plt.ylabel("No.Of Students")
plt.show()
```

```python
# Bre Chart :-
```

```python
plt.figure(figsize = (12 , 2))
sns.barplot(x = "Course" , y = "Fees" , data  = df , estimator = len)
plt.title("Total Sales by Course")
plt.xlabel("Course")
plt.show()
```

```python
plt.figure(figsize = (6 , 2))
sns.histplot(x = "Age" , data = df , bins = 6)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

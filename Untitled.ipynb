```python
def prime(num):
    count = 0
    if type(num) == int or type(num) == float:
        for i in range (num):
            count += 1
        return count
```

```python
print(prime(10))
print(prime(123,45))
print(prime([10,20,30,40]))
```

# Function assignment :-

```python
def reverse_string(s):
    return s[::-1]
print(reverse_string("hello"))
```

```python
def count_vowels(s):
    vowels = "aeiou"
    count = 0
    for ch in s:
        if ch in vowels:
            count +=1
    return count
print(count_vowels("aeiou"))
```

```python
def is_prime (num):
    if num <= 1:
        return false
    for i in range(2, int(num**0.5)+ 1):
        if num%i==0:
            return False
    return True
print(is_prime(29))
```

```python
def largest_el_list(list):
    return max(list)
print(largest_el_list([3,7,2,9,5]))
```

```python
def remove_duplicates(lst):
    return list(set(lst))
print(remove_duplicates([1,2,2,3,4]))
```

```python
def factorial(n):
    fact =  1
    for i in range(1, n+1):
        fact *= i
    return fact
print(factorial(5))
```

```python
def is_palindrome(string):
    return string == string[::-1]
print(is_palindrome("python"))
```

```python
def fibonacci(n):
    seq = [0, 1]
    while len(seq) < n:
        seq.append(seq[-1] + seq [-2])
    return seq[:n]
print(fibonacci(7))
```

```python
def merge_sorted_lists(l1,l2):
    return sorted(l1 + l2)
print(merge_sorted_lists([1,3,5],[2,4,6]))
```

```python
def second_largest(lst):
    num = list(set(lst))
    num.sort()
    return num[-2]
print(second_largest([10,20,4,45,99]))
```

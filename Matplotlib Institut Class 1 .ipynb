# Pie Chart

```python
plt.figure(figsize = (6,2))
gender = ["Male", "Female"]
total_students = [730,380]

plt.pie(total_students , labels = gender , autopct = "%.f%%" , colors = ["skyblue" , "Yellow"])
plt.title("Totale Students By Gender")

plt.show()
```

```python
plt.figure(figsize = (6,2))
gender = ["Male", "Female"]
total_students = [730,380]

plt.pie(total_students , labels = gender ,radius = 1 , autopct = "%.f%%" , colors = ["skyblue" , "Yellow"] , textprops = {
    "fontsize" : 10 , "color" : "red" , "weight" : "bold"
})
plt.title("Total Students By Gender")

plt.show()
```

```python
plt.figure(figsize = (6,3))
course = ["Data Analytics" , "Data Science" , "AI Development"]
Sales = [184,133,59]
e = [0,0,0.3]
plt.pie(Sales , labels = course , autopct = "%.f%%" , radius = 1 , explode = e)
plt.title("Total Students By Gender")

plt.show()
```

# Subplots:-

```python
import matplotlib.pyplot as plt
import numpy as np

# Axes = ye actual graph ka area hot hai jisme :- line banti hai , bar banta hai , pie chart banti hai
# or ye hi yiska formula fig , axes = plt.subplots(2,2)..


fig , axes = plt.subplots(2,2 , figsize = (7 , 5))
year = ["2022","2023","2024" ,"2025"]
Sales = [48 , 52, 72, 65]
axes[0,0].plot(year , Sales)
axes[0,0].set_title("Total Sales Over Years")
axes[0,0].set_xlabel("Year")
axes[0,0].set_ylabel("Sales")


Product = ["Laptop" , "Mobile" , "TV" , "AC"]
sales = [78 ,62, 40, 24]
axes[0,1].bar(Product , sales , color = ["blue" , "black" ,"yellow"  , "green"])
axes[0,1].set_title("Sales by Product" , color = "red")
axes[0,1].set_xlabel("Product" , color = "green")
axes[0,1].set_ylabel("sales")


gender = ["Male" , "Female"]
sales= [78 , 22]
axes[1,0].pie(sales , labels = gender , autopct = "%.f%%")
axes[1,0].set_title("Total Sales by Gender")
axes[1,0].set_xlabel("gender")
axes[1,0].set_ylabel("sales")


sales =  np.random.rand(100)*100
profit = np.random.rand(100)*10
axes[1,1].scatter(sales , profit , marker = "*")
axes[1,1].set_title("Sales  vs Profit")
axes[1,1].set_xlabel("Sales")
axes[1,1].set_ylabel("profit")


plt.tight_layout()
plt.show()
```

```python
y23 = df[df["Enrollment Date"].dt.year ==2023]
y24 = df[df["Enrollment Date"].dt.year ==2024]
y25 = df[df["Enrollment Date"].dt.year ==2025]
```

```python
p23 = y23.groupby("Placement Status")["Name"].count()
p23 = list(p23)

p24 = y24.groupby("Placement Status")["Name"].count()
p24 = list(p24)

p25 = y25.groupby("Placement Status")["Name"].count()
p25 = list(p25)
```

```python
fig , ax = plt.subplots(1,3 ,figsize = (10 ,3))

ax[0].pie(p23 , labels = ["Not Placed" , "Placed"] , autopct = "%.f%%")
ax[0].set_title("Placement Status(2023)")

ax[1].pie(p24 , labels = ["Not Placed" , "Placed"] , autopct = "%.f%%")
ax[1].set_title("Placement Status(2024)")

ax[2].pie(p25 , labels = ["Not Placed" , "Placed"] , autopct = "%.f%%")
ax[2].set_title("Placement Status(2025)")

plt.tight_layout()
plt.show()
```

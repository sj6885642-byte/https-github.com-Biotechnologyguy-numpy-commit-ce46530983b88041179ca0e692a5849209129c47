```python
import matplotlib.pyplot as plt
plt.figure(figsize = (6,2))
x = ["2021" , "2022", "2023" , "2024" , "2025", ]
y = [44 , 57 , 48 , 62 ,74]
plt.plot(x,y,color = "blue" , linewidth = 1.5 , marker = "o" , linestyle = "--")
plt.title("Total Sales by Year")
plt.show()
```

```python
plt.figure(figsize = (6,2))
x = ["2021" , "2022", "2023" , "2024" , "2025", ]
y = [44 , 57 , 48 , 62 ,74]
plt.plot(x,y,color = "blue" , linewidth = 1.5 , marker = "o" , linestyle = "--" , markersize = 10)
plt.title("Total Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize = (6,2))
x = ["2021" , "2022", "2023" , "2024" , "2025", ]
y = [44 , 57 , 48 , 62 ,74]
plt.plot(x,y,color = "blue" , label = "Sales")
plt.title("Total Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales")


plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (6,2))
year= ["2021" , "2022", "2023" , "2024" , "2025", ]
apple = [44 , 57 , 48 , 62 ,74]
samsung = [33,48,54,50,66]

plt.plot(year, apple , color = "blue" , label = "Apple")
plt.plot(year , samsung , color = "red" , label  = "Samsung")
plt.title("Total Sales By Year(Apple vs Samsung)" , fontsize = 15)
plt.xlabel("Year" , fontsize = 18)
plt.ylabel("Sales" , fontsize = 18)


plt.grid(True)
plt.legend()
plt.show()
```

# pie Chart :-

```python
plt.figure(figsize = (6,2))
gender = ["Male", "Female"]
total_students = [730,380]

plt.pie(total_students , labels = gender , autopct = "%.2f%%" , colors = ["skyblue" , "Yellow"])
plt.title("Totale Students By Gender")

plt.show()
```

```python
plt.figure(figsize = (6,2))
gender = ["Male", "Female"]
total_students = [730,380]

plt.pie(total_students , labels = gender ,radius = 1 , autopct = "%.f%%" , colors = ["skyblue" , "Yellow"] , textprops = {
    "fontsize" : 10 , "color" : "k" , "weight" : "bold"
})
plt.title("Total Students By Gender")

plt.show()
```

```python
plt.figure(figsize = (6,2))
course = ["Data Analytics" , "Data Science" , "AI Development"]
Sales = [184,133,59]
e = [0,0,0.3]
plt.pie(Sales , labels = course , autopct = "%.f%%" , radius = 0.8 , explode = e)
plt.title("Total Students By Gender")

plt.show()
```
